```python
# importing major libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('placement.csv')
```

```python
plt.figure(figsize=(12,5))
plt.scatter(df.cgpa,df.package)
plt.title('CGPA VS Package')
plt.xlabel('CGPA')
plt.ylabel('PKG(In LPA)')
plt.show()
```

```python
df.corr()
```

```python
# x --> independent variable // input
# y ---> dependent variable // output

x= df.iloc[:,0:1]
y= df.package # 1-d/Series
x
```

## Train–Test Split Workflow

### Training Phase
- X_train → Input features for training
- y_train → Target values for training
- fit() → Model learns from X_train and y_train

### Testing Phase
- X_test → Unseen input data
- predict() → Model generates predictions

### Evaluation / Comparison
- Predictions → Output predicted by the model
- y_test → Actual (original) output
- Compare predictions with y_test using evaluation metrics

### Flow
X_train + y_train → fit() → trained model  
X_test → predict() → predictions  
predictions ↔ y_test → evaluation

```python
#ml --> scikit learn
# model_selection --> train_test_split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
#x_train,y_train
print('rows for training',x_train.shape[0])
print('rows for testing',x_test.shape[0])
```

```python
#algorithm import
#hero
#linear regression

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr
```

```python
# training
# .fit --> method
lr.fit(x_train,y_train)
```

```python
#questioning
#testing

y_pred = lr.predict(x_test)
```

```python
# actual result
# y_test
y_test.iloc[0],y_pred[0]
y_test.iloc[1],y_pred[1]
```

```python
#Best fit line
plt.figure(figsize=(12,5))
plt.scatter(df.cgpa,df.package,label='test')
plt.scatter(x_train,y_train,label='train')
plt.plot(x_test,lr.predict(x_test),color='r',label='Best fit line')
plt.title('CGPA VS Package')
plt.xlabel('CGPA')
plt.ylabel('PKG(In LPA)')
plt.show()
```

```python
y_test.iloc[0],y_pred[0]
y_test.iloc[1],y_pred[1]
#y_test-y_pred
```

```python
#metrics
#regression --> r2_score,mae,mse,rmse
```

```python
#r2_score/coefficient of determination/goodness of fit
#overall performance of model
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
